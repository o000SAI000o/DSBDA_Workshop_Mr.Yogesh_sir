**Exercise 1: Basic Plot**

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.random.rand(20)
y = x * 2

fig = plt.figure()
ax = fig.add_axes([0,0,1,1])

ax.plot(x,y,'b')
ax.set_xlabel("X Label")
ax.set_ylabel("Y Label")
ax.set_title("Basic Plot")

plt.show()
```

**Exercise 2: Two Subplots**

```python
fig = plt.figure()

ax1 = fig.add_axes([0,0,1,1])
ax1.plot(x,y,'r')
ax1.set_title("Larger plot")

ax1 = fig.add_axes([0.2,0.5,0.4,0.4])
ax1.plot(x,y,'g')
ax1.set_title("Smaller plot")

plt.show()
```

 **Exercise 3: Subplots with plt.subplots()**

```python
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(10, 4))

# First subplot
axes[0].plot(x, y, 'b')
axes[0].set_title("First Plot")

# Second subplot
axes[1].plot(x, x**2, 'r')
axes[1].set_title("Second Plot")

plt.show()

```

**Exercise 4: Customization**

```python
fig, ax = plt.subplots(figsize=(8, 6))

ax.plot(x, y, color="purple", lw=3, ls="--", marker="o", markersize=5, markerfacecolor="yellow")

ax.set_xlabel("X Label")
ax.set_ylabel("Y Label")
ax.set_title("Styled Plot")

plt.show()
```

**Exercise 5: Legends & Grid**

```python
fig, ax = plt.subplots()

ax.plot(x, y, label="y = 2x")
ax.plot(x, x**2, label="y = x^2")

ax.legend()
ax.grid(True)
ax.set_title("Legend & Grid Example")

plt.show()
```
